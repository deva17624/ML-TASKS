###  K-Nearest Neighbors(Knn)

 K-Nearest Neighbors is a supervised machine learning algorithm used for both classification and regression problems.

 Knn is predictions a new data point or target value  by comparing the data pionts with the k most similar data points from the train  dataset or data pionts.

 The final output is determined based on distance-based similarity.

Closer neighbors → higher influence

Farther neighbors → lower influence

**k values**

k is values tells how many pionts to are taken k-neghbors like k=1 ork=2 ....

k (n_neighbors): Number of neighbors

**Types**



IN Knn problem two types : 
> 1- classification
 
> 2- regression 

For classfiction we use mode  for k values 

For regresion we use mean  k values 

**Feature Scaling**

KNN is distance-based algorthim , so features with larger scales dominate the distance calculation.

- for example, Age range is  0–100 and Income range is  0–10,00,000 if go Without scaling, income will have more influence age.

so we use Normalization (Min–Max scaling and etc)abs 

- For categrocial data we cannot find the distance. so we use Onehotencoding  

**Distances**

 For distances calculation
 
 Euclidean Distance
> Straight-line distance; most commonly used.

Manhattan Distance

Minkowski Distance

Cosine Similarity


**Inner logic**

- In training this algorithm is do nothing  just store the data

- In testing, each testing pionts are comparing all the training data pionts which is simmilar or near using euclidean distances etc.

- based on the  k values  select the k neighbours  values

- for each testing piont we find k no of values 

- if it classfiction we use mode on  k values or if  regresion we use mean on k values


**Advantages and Limitations**


> Advantages

Simple and easy to understand

No training phase 

Works well with small to medium datasets

Flexible for both classification and regression

> Limitations

Slow prediction for large datasets

High memory usage because stores all data

Sensitive to irrelevant features

Performance heavily depends on feature scaling and k

```python
import pandas as pd 
import numpy as np 
```

```python
df=pd.read_csv(r"C:\Users\devav\Downloads\diamonds.csv")
df.head()
```

```python
x=df.drop("price",axis=1)
x.head()
```

```python
y=df["price"]
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_cols = ["carat", "depth", "table", "x", "y", "z"]
cat_cols = ["cut", "color", "clarity"] 
preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols)
    ]
)
```

```python
x_train_= preprocessor.fit_transform(x_train)
x_train_
```

```python
x_test_ = preprocessor.transform(x_test)
x_test_
```

```python
from sklearn.neighbors import KNeighborsRegressor

knn_sklearn = KNeighborsRegressor(n_neighbors=3)
knn_sklearn.fit(x_train_, y_train)
y_pred_sklearn = knn_sklearn.predict(x_test_)
```

```python
from sklearn.metrics import  r2_score

r2_sklearn = r2_score(y_test, y_pred_sklearn)

print("Sklearn KNN R2:", r2_sklearn)
```
